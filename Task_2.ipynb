```python
survived = titanic_data['Survived'].value_counts()
labels = survived.index
values = survived.values
```

```python
data = go.Pie(labels = labels,
      values = values,
      name = 'Survival Status',
      marker = dict(colors = ['rose', 'lightblue']))

layout = go.Layout(title = 'Survival Status')

fig = go.Figure(data = [data], layout = layout)
iplot(fig)
```

```python
fig = px.scatter(titanic_data, x='Fare', y='Age', color='Survived', size='Fare')


fig.show()
```

```python
surv = titanic_data.loc[titanic_data['Survived'] == 1, :]
n_surv = titanic_data.loc[titanic_data['Survived'] == 0, :]
```

```python
trace1 = go.Box(y = surv['Age'],
             name = 'Survived',
             marker = dict(color = 'blue'))

trace2 = go.Box(y = n_surv['Age'],
             name = 'Not Survived',
             marker = dict(color = 'green'))

layout = go.Layout(title = 'Survival by Age Group',
                  xaxis = dict(title = 'Survival Status'),
                  yaxis = dict(title = 'Age'))

fig = go.Figure(data = [trace1, trace2], layout = layout)
iplot(fig)
```

```python
surv_gender = surv['Sex'].value_counts()
n_surv_gender = n_surv['Sex'].value_counts()
```

```python
fig = make_subplots(rows = 1, cols = 2, specs = [[{'type' : 'domain'}, {'type' : 'domain'}]])

fig.add_trace(go.Pie(labels = surv_gender.index,
       values = surv_gender.values,
       name = 'Survived',
       marker = dict(colors = ['lightblue', 'lightgreen'])), 1, 1)

fig.add_trace(go.Pie(labels = n_surv_gender.index,
       values = n_surv_gender.values,
       name = 'Not Survived',
       marker = dict(colors = ['lightblue', 'lightgreen'])), 1, 2)

fig.update_traces(hole = 0.3)

fig.update_layout(title_text = 'Survival by Gender',
                 annotations = [dict(text = 'Survived', showarrow = False, x = 0.19),
                            dict(text = 'Not Survived', showarrow = False, x = 0.82)])



fig.show()
```

```python
surv_sib = titanic_data.loc[titanic_data['Survived'] == 1, :]
n_surv_sib = titanic_data.loc[titanic_data['Survived'] == 0, :]
```

```python
fig = make_subplots(rows = 1, cols = 2)

fig.add_trace(go.Histogram(x = surv_sib['SibSp'],
            name = 'Survived',
            opacity = 0.8,
            marker = dict(color = 'teal')), 1, 1)

fig.add_trace(go.Histogram(x = n_surv_sib['SibSp'],
            name = 'Not Survived',
            opacity = 0.8,
            marker = dict(color = 'tan')), 1, 2)

fig.update_layout(title_text = 'Survival by Siblings/Spouse Aboard',)

fig.show()
```

```python
fig = make_subplots(rows = 1, cols = 2)

fig.add_trace(go.Histogram(x = surv_sib['Parch'],
            name = 'Survived',
            opacity = 0.8,
            marker = dict(color = 'tan')), 1, 1)

fig.add_trace(go.Histogram(x = n_surv_sib['Parch'],
            name = 'Not Survived',
            opacity = 0.8,
            marker = dict(color = 'teal')), 1, 2)

fig.update_layout(title_text = 'Survival by Parents/Children Aboard')

fig.show()
```
